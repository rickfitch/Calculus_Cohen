Calculus formula for e = 2.718
e=(1+(1/n))**n
Whereas n = the set of real numbers

```python
n = [1,2,5,10,20]
print(f"e = {np.e:.4f}")

for i in n:
    e = (1 + (1/i)) ** i
    print(f"n = {i:2.0f}, e = {e:.4f}, diff = {abs(e - np.e):.4f}")
   
```

```python
n = np.arange(1, 1001)
eDiff = np.e -(1 + (1/n)) ** n 
plt.plot(n, eDiff)
plt.grid()
plt.title("Difference between e and (1 + 1/n)^n")
```

Various natural log functions

```python
xDomain = [-2, 2]
yDomain = [-0.5, 4]
plt.xlim(xDomain)
x = np.linspace(xDomain[0], xDomain[1], 53)
plt.ylim(yDomain)

y1 = np.exp(x)
y2 = np.exp(x**2)
y3 = np.exp(-x)**2
y4 = np.exp(-x**2)
y5 = np.exp(x)**2

plt.plot(x, y2, label='exp(x)')
plt.plot(x, y3, label='exp(-x)^2')
plt.plot(x, y4, label='exp(-x^2)')
plt.plot(x, y5, label='exp(x)^2')
plt.legend()
plt.grid()
plt.title('Exponential Functions')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
```
